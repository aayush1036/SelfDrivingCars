```python
import pandas as pd 
import os 
import cv2 
import matplotlib.pyplot as plt 
```

```python
TRAIN_LABELS = os.path.join('bdd100k','labels', 'bdd100k_labels_images_train.json')
VAL_LABELS = os.path.join('bdd100k','labels', 'bdd100k_labels_images_val.json')
TRAIN_IMAGES = os.path.join('bdd100k','images','100k','train')
VAL_IMAGES = os.path.join('bdd100k','images','100k', 'val')
```

```python
train_labels = pd.read_json(TRAIN_LABELS)
val_labels = pd.read_json(VAL_LABELS)
```

```python
train_sample = train_labels.sample(n=5)
```

```python
for _, row in train_sample.iterrows():
    _, ax = plt.subplots(figsize=(8,8))
    img = cv2.imread(os.path.join(TRAIN_IMAGES, row['name']))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    attr = row['attributes']
    label = [f'{k}-{v}' for k, v in attr.items()]
    label_text = '\n'.join(label)
    
    labels = row['labels']
    for label in labels: 
        if 'box2d'  in label.keys():
            cat = label['category']
            pt1 = (int(label['box2d']['x1']), int(label['box2d']['y1']))
            pt2 = (int(label['box2d']['x2']), int(label['box2d']['y2']))
            img = cv2.rectangle(img, pt1, pt2, color=(0,255,0), thickness=2)
            img = cv2.putText(img, cat, pt1, fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=1, color=(0,255,0))
        else:
            continue
    ax.axis('off')
    ax.imshow(img)
    ax.set_title(label_text)
    plt.show()
```
